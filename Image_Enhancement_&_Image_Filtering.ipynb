# Image Enhancement

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Load the input image

```python
input_image = cv2.imread("IMG_7387.JPG", cv2.IMREAD_GRAYSCALE)
plt.imshow(input_image, cmap='gray')
```

Normalize the image to the range [0, 1]


```python
normalized_image = input_image / 255.0
```

Apply the gamma correction

```python
enhanced_image = np.power(normalized_image, 1.5)

#power  =>  ()
#power  =>  ()
```

Scale back to the range [0, 255]

```python
enhanced_image = (enhanced_image * 255).astype(np.uint8)

plt.imshow(enhanced_image, cmap='gray')
plt.show()
```

# Image Filtering

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the input image
input_image = cv2.imread("IMG_7387.JPG", cv2.IMREAD_GRAYSCALE)
plt.imshow(input_image, cmap='gray')
```

1. Blurring (Averaging)


Simply takes the average of all the pixels in the kernel area

```python
blurred = cv2.blur(input_image, (5, 5))  # Kernel size = 5x5
plt.imshow(blurred, cmap='gray')
```

```python
blurred = cv2.blur(input_image, (25, 25))
plt.imshow(blurred, cmap='gray')
```

2. Gaussian Blurring

Gaussian blurring is highly effective in removing gaussian noise from the image.

```python
gaussian_blurred = cv2.GaussianBlur(input_image, (5, 5), 0)  # Kernel size = 5x5, sigma(S.D =>   ) = 0
plt.imshow(gaussian_blurred, cmap='gray')
```

3. Median Blurring => Good for Salt and Pepper

This is highly effective against salt-and-pepper noise in an image.

```python
median_blurred = cv2.medianBlur(input_image, 5)  # Kernel size = 5 (must be an odd number)
plt.imshow(median_blurred, cmap='gray')
```

4. Sharpening (Enhancing edges)

```python
# Define the sharpening kernel
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])
sharpened = cv2.filter2D(input_image, -1, sharpen_kernel)
plt.imshow(sharpened, cmap='gray')
```
